```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
import numpy as np
import re

def get_feature_names(text):
    clean_text = [re.sub(r"\d+", "",x) for x in text]
    cv = CountVectorizer(stop_words=all_stopwords)   
    cv_fit = cv.fit_transform(clean_text)    
    word_list = cv.get_feature_names_out() 
    return word_list
    
def get_count_list(text):
    clean_text = [re.sub(r"\d+", "",x) for x in text]
    cv = CountVectorizer(stop_words=all_stopwords)   
    cv_fit = cv.fit_transform(clean_text)    
    word_list = cv.get_feature_names_out() 
    count_list = np.asarray(cv_fit.sum(axis=0))
    counts = dict(zip(word_list, count_list[0]))
    sorted_counts = {k: v for k, v in sorted(counts.items(), key=lambda item: item[1], reverse = True)}
    return sorted_counts
```

```python
def get_processable_references(reference):
    processable_references = []
    for x in reference:
        if 'author' in x:
            processable_references.append(x['author'])
        elif 'unstructured' in x:
            processable_references.append(x['unstructured'])
    return processable_references

reference_df['processable_references'] = reference_df['reference'].apply(lambda x: get_processable_references(x))
```

```python
reference_df['processable_ref_count'] = reference_df['processable_references'].apply(lambda x: len(x))
```

```python
reference_df['percentage'] = round((reference_df['processable_ref_count']/reference_df['reference-count'])* 100)
```
